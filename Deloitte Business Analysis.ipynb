```python
print("DELOITTE BUSINESS DATASET OVERVIEW:-")


print("\nTotal Employees :", df['Employee_Count'].sum())

print("\nTotal Revenue : ₹{:,.0f}".format(df['Monthly_Revenue'].sum()))

print("\nTotal Expenses : ₹{:,.0f}".format(df['Monthly_Expenses'].sum()))

print("\nTotal Profit : ₹{:,.0f}".format(df['Profit'].sum()))

print("\nAverage Client Satisfaction : {:.2f}".format(df['Client_Satisfaction'].mean()))

print("\nAverage Employee Productivity : {:.2f}%".format(df['Employee_Productivity'].mean()))
```

```python
#Total Monthly Revenue
total_revenue = df['Monthly_Revenue'].sum()
print("Total Monthly Revenue: ₹{:,.0f}".format(total_revenue))
```

```python
#Total Monthly Expenses
total_expenses = df['Monthly_Expenses'].sum()
print("Total Monthly Expenses: ₹{:,.0f}".format(total_expenses))
```

```python
#Total Profit
total_profit = df['Profit'].sum()
print("Total Profit: ₹{:,.0f}".format(total_profit))
```

```python
#Average Client Satisfaction
avg_satisfaction = df['Client_Satisfaction'].mean()
print("Average Client Satisfaction :", round(avg_satisfaction,2))
```

```python
#Average Employee Productivity
avg_productivity = df['Employee_Productivity'].mean()
print("Average Employee Productivity :", round(avg_productivity,2),"%")
```

```python
#Departmet-wise Revenue
dept_revenue = df.groupby('Department')['Monthly_Revenue'].sum()
print(dept_revenue)
```

```python
#Departmet-wise Profit
dept_profit = df.groupby('Department')['Profit'].sum()
print(dept_profit)
```

```python
#Region-wise Revenue
region_revenue = df.groupby('Region')['Monthly_Revenue'].sum()
print(region_revenue)
```

```python
#Branch-wise Revenue
branch_revenue = df.groupby('Branch')['Monthly_Revenue'].sum()
print(branch_revenue)
```

```python
#Quarter-wise Revenue
quarter_revenue = df.groupby('Quarter')['Monthly_Revenue'].sum()
print(quarter_revenue)
```

```python
#Year-wise Revenue
year_revenue = df.groupby('Year')['Monthly_Revenue'].sum()
print(year_revenue)
```

```python
#Highest Revenue Department
highest_department = dept_revenue.idxmax()

print("Department with Highest Revenue :", highest_department)
```

```python
#Highest Profit Branch
highest_branch = branch_revenue.idxmax()
print("Branch with Highest Revenue: ", highest_branch)
```

```python
correlation = df.corr(numeric_only=True)
print(correlation)
```

```python
#Visualization 1: Department-wise Revenue
plt.figure(figsize=(10,6))

dept_revenue.plot(kind='bar',color='gold')

plt.title("Department-wise Revenue")

plt.xlabel("Department")

plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.grid(axis='y')

plt.show()
```

```python
#Visualization 2: Department-wise Profit 
plt.figure(figsize=(10,6))

dept_profit.plot(kind='bar',color='darkmagenta')

plt.title("Department-wise Profit")

plt.xlabel("Department")

plt.ylabel("Profit")

plt.xticks(rotation=45)

plt.grid(axis='y')

plt.show()
```

```python
#Visualization 3: Region-wise Revenue
plt.figure(figsize=(8,6))

region_revenue.plot(kind='pie',autopct='%1.1f%%')

plt.ylabel("")

plt.title("Region-wise Revenue")

plt.show()
```

```python
#Visualization 4: Revenue vs Expenses
plt.figure(figsize=(8,5))

plt.scatter(df['Monthly_Revenue'],df['Monthly_Expenses'])

plt.xlabel("Revenue")

plt.ylabel("Expenses")

plt.title("Revenue vs Expenses")

plt.grid()

plt.show()
```

```python
#Visualization 5: Profit Distribution
plt.figure(figsize=(8,5))

plt.hist(df['Profit'],bins=20)

plt.title("Profit Distribution")

plt.xlabel("Profit")

plt.ylabel("Frequency")

plt.show()
```

```python
#Visualization 6: Employee Productivity
plt.figure(figsize=(8,5))

plt.hist(df['Employee_Productivity'],bins=15)

plt.title("Employee Productivity Distribution")

plt.xlabel("Productivity (%)")

plt.ylabel("Employees")

plt.show()
```

```python
#Visualization 7: Client Satisfaction
plt.figure(figsize=(8,5))

plt.hist(df['Client_Satisfaction'],bins=10)

plt.title("Client Satisfaction")

plt.xlabel("Rating")

plt.ylabel("Frequency")

plt.show()
```

```python
#Visualization 8: Project Count by Department
project = df.groupby('Department')['Project_Count'].sum()

plt.figure(figsize=(10,6))

project.plot(kind='bar',color='mediumaquamarine')

plt.title("Projects Completed by Department")

plt.ylabel("Projects")

plt.grid(axis='y')

plt.show()
```

```python
#Visulization 9: Training Hours by Department
training = df.groupby('Department')['Training_Hours'].mean()

plt.figure(figsize=(10,6))

training.plot(kind='bar',color='rebeccapurple')

plt.title("Average Training Hours")

plt.ylabel("Hours")

plt.grid(axis='y')

plt.show()
```

```python
#Visualization 10: Average Revenue by Year
year_avg = df.groupby('Year')['Monthly_Revenue'].mean()

plt.figure(figsize=(8,5))

year_avg.plot(marker='o')

plt.title("Average Revenue by Year")

plt.xlabel("Year")

plt.ylabel("Revenue")

plt.grid()

plt.show()
```
